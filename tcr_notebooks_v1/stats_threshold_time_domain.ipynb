**stats_threshold_time_domain.ipynb**

Function: given F1 / IOU scores at different aoi, frame, threshold, calculate threshold with largest F1 / IOU (to see whether prediction is stable) 

Input: segmentation_infer_time_domain_tcr_v1_alpha_05_modified.csv / segmentation_infer_time_domain_tcr_v1_alpha_09_modified.csv / segmentation_infer_time_domain_tcr_v1_alpha_1_modified.csv

```python
import pandas as pd
import os, glob
```

```python
path = "segmentation_infer_time_domain/"
all_files = glob.glob(os.path.join(path, "segmentation_infer_time_domain_tcr_v1_alpha_1*.csv"))
print(all_files)
df_from_each_file = (pd.read_csv(f, sep=',') for f in all_files)
df_merged = pd.concat(df_from_each_file, ignore_index=True)
df_merged.to_csv("segmentation_infer_time_domain/segmentation_infer_time_domain_tcr_v1_alpha_1.csv")
```

```python
def stats_threshold_time_domain(file, outfile):
    # file -> editable
    df = pd.read_csv(file)

    aois = df['AOI'].unique()
    print(sorted(aois))
    
    new_df = pd.DataFrame()

    for aoi in aois:
        print(aoi)

        tmp = pd.DataFrame(df)[df['AOI'] == aoi]
        frames = tmp['Frame'].unique()

        for frame in frames:
            print(frame)
            tmptmp = pd.DataFrame(tmp)[tmp['Frame'] == frame]

            if (tmptmp.Dice.max() == 0.0):
                continue

            print(tmptmp[tmptmp.Dice == tmptmp.Dice.max()])
            new_df = pd.concat([new_df, tmptmp[tmptmp.Dice == tmptmp.Dice.max()]])

    # outfile -> editable
    new_df.to_csv(outfile, index=False, header=True)
```

```python
file = 'segmentation_infer_time_domain/segmentation_infer_time_domain_tcr_v1_alpha_05.csv'
outfile = 'segmentation_infer_time_domain/stats_threshold_time_domain_05.csv'
stats_threshold_time_domain(file, outfile)
```

```python
file = 'segmentation_infer_time_domain/segmentation_infer_time_domain_tcr_v1_alpha_09.csv'
outfile = 'segmentation_infer_time_domain/stats_threshold_time_domain_09.csv'
stats_threshold_time_domain(file, outfile)
```
